# Librerías y arreglo de tiempo 

```python
# Importamos las librerías que se van a utilizar
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, gamma, lognorm, expon, beta

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses
tiempos = np.arange(0, 12.01, 0.01)
```

# Motor Diesel 1 

### Componente D1

```python
# Parámetros de la distribución uniforme
a = 3.7  
b = 6.0  

# Crear una instancia de la distribución uniforme
dist_uniforme_D1 = uniform(loc=a, scale=b-a)

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses
fiabilidad_D1 = 1 - dist_uniforme_D1.cdf(tiempos)

# Imprimir los valores de fiabilidad
for tiempo, fiabilidad in zip(tiempos, fiabilidad_D1):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad: {fiabilidad:.4f}")
```

### Componente C3

```python
# Parámetros de la distribución gamma
shape = 8  
scale = 3  

# Crear una instancia de la distribución gamma
dist_gamma_C3 = gamma(a=shape, scale=scale)

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses
fiabilidad_C3 = 1 - dist_gamma_C3.cdf(tiempos)

# Imprimir los valores de fiabilidad
for tiempo, fiabilidad in zip(tiempos, fiabilidad_C3):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad: {fiabilidad:.4f}")
```

### BLOQUE A

```python
# Cálculo de la fiabilidad del bloque "a" para cada valor de tiempo entre 0 y 12 meses
fiabilidad_bloque_a = fiabilidad_D1 * fiabilidad_C3

# Imprimir los valores de fiabilidad del bloque "a"
for tiempo, fiabilidad in zip(tiempos, fiabilidad_bloque_a):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad del bloque 'a': {fiabilidad:.4f}")
```

### Componente E5

```python
# Parámetros de la distribución lognormal para E5
shape_E5 = 2
scale_E5 = 1.2

# Crear una instancia de la distribución lognormal para E5
dist_lognormal_E5 = lognorm(s=shape_E5, scale=scale_E5)

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses para E5
fiabilidad_E5 = 1 - dist_lognormal_E5.cdf(tiempos)

# Imprimir los valores de fiabilidad para E5
for tiempo, fiabilidad in zip(tiempos, fiabilidad_E5):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad de E5: {fiabilidad:.4f}")
```

### Componente E7

```python
# Parámetros de la distribución lognormal para E7
shape_E7 = 1
scale_E7 = 1.1

# Crear una instancia de la distribución lognormal para E7
dist_lognormal_E7 = lognorm(s=shape_E7, scale=scale_E7)

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses para E7
fiabilidad_E7 = 1 - dist_lognormal_E7.cdf(tiempos)

# Imprimir los valores de fiabilidad para E7
for tiempo, fiabilidad in zip(tiempos, fiabilidad_E7):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad de E7: {fiabilidad:.4f}")
```

### BLOQUE B

```python
# Cálculo de la fiabilidad del bloque "b" para cada valor de tiempo entre 0 y 12 meses
fiabilidad_bloque_b = 1 - (1 - fiabilidad_E5) * (1 - fiabilidad_E7)

# Imprimir los valores de fiabilidad del bloque "b"
for tiempo, fiabilidad in zip(tiempos, fiabilidad_bloque_b):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad del bloque 'b': {fiabilidad:.4f}")
```

### Componente C9

```python
# Parámetros de la distribución gamma para C9
shape_C9 = 2
scale_C9 = 2

# Crear una instancia de la distribución gamma para C9
dist_gamma_C9 = gamma(a=shape_C9, scale=scale_C9)

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses para C9
fiabilidad_C9 = 1 - dist_gamma_C9.cdf(tiempos)

# Imprimir los valores de fiabilidad para C9
for tiempo, fiabilidad in zip(tiempos, fiabilidad_C9):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad de C9: {fiabilidad:.4f}")
```

### BLOQUE H

```python
# Cálculo de la fiabilidad del bloque "h" para cada valor de tiempo entre 0 y 12 meses
fiabilidad_bloque_h = fiabilidad_C9

# Imprimir los valores de fiabilidad del bloque "h"
for tiempo, fiabilidad in zip(tiempos, fiabilidad_bloque_h):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad del bloque 'h': {fiabilidad:.4f}")
```

### Componente H15

```python
# Parámetro de la distribución exponencial para H15
scale_H15 = 10

# Crear una instancia de la distribución exponencial para H15
dist_expon_H15 = expon(scale=scale_H15)

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses para H15
fiabilidad_H15 = dist_expon_H15.sf(tiempos)

# Imprimir los valores de fiabilidad para H15
for tiempo, fiabilidad in zip(tiempos, fiabilidad_H15):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad de H15: {fiabilidad:.4f}")
```

### BLOQUE J

```python
# Cálculo de la fiabilidad del bloque "j" para cada valor de tiempo entre 0 y 12 meses
fiabilidad_bloque_j = fiabilidad_H15

# Imprimir los valores de fiabilidad del bloque "j"
for tiempo, fiabilidad in zip(tiempos, fiabilidad_bloque_j):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad del bloque 'j': {fiabilidad:.4f}")
```

### MOTOR DIESEL 1 (BLOQUE A + BLOQUE B + BLOQUE H + BLOQUE J)

```python
# Cálculo de la fiabilidad del Motor Diesel 1 para cada valor de tiempo entre 0 y 12 meses
fiabilidad_motor_diesel_1 = fiabilidad_bloque_a * fiabilidad_bloque_b * fiabilidad_bloque_h * fiabilidad_bloque_j

# Imprimir los valores de fiabilidad del Motor Diesel 1
for tiempo, fiabilidad in zip(tiempos, fiabilidad_motor_diesel_1):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad del Motor Diesel 1: {fiabilidad:.4f}")
```

# Turbina

### Componente T11

```python
# Parámetros de la distribución beta para T11
a_T11 = 5
b_T11 = 1
loc_T11 = 2
scale_T11 = 12

# Crear una instancia de la distribución beta para T11
dist_beta_T11 = beta(a=a_T11, b=b_T11, loc=loc_T11, scale=scale_T11)

# Cálculo de la fiabilidad para cada valor de tiempo entre 0 y 12 meses para T11
fiabilidad_T11 = 1 - dist_beta_T11.cdf(tiempos)

# Imprimir los valores de fiabilidad para T11
for tiempo, fiabilidad in zip(tiempos, fiabilidad_T11):
    print(f"Tiempo: {tiempo:.2f} meses, Fiabilidad de T11: {fiabilidad:.4f}")
```

# COMBINACIONES 

### Motor Diesel 1 + Motor Diesel 2 + Turbina

```python
# Calcular la probabilidad de fiabilidad conjunta del Motor Diesel 1, Motor Diesel 2 y Turbina
probabilidad_diesel1_diesel2_turbina = fiabilidad_motor_diesel_1 + fiabilidad_motor_diesel_2 + fiabilidad_turbina - \
                                  fiabilidad_motor_diesel_1 * fiabilidad_motor_diesel_2 - \
                                  fiabilidad_motor_diesel_1 * fiabilidad_turbina - \
                                  fiabilidad_motor_diesel_2 * fiabilidad_turbina + \
                                  fiabilidad_motor_diesel_1 * fiabilidad_motor_diesel_2 * fiabilidad_turbina

# Imprimir los valores de la probabilidad de fiabilidad conjunta
for tiempo, probabilidad in zip(tiempos, probabilidad_diesel1_diesel2_turbina):
    print(f"Tiempo: {tiempo:.2f} meses, Probabilidad de Fiabilidad conjunta del Motor Diesel 1, Motor Diesel 2 y Turbina: {probabilidad:.4f}")
```
